```python
from pdfminer.high_level import extract_text

resume_text = extract_text("Gaurav_cv.pdf")  # replace with your filename
doc = nlp(resume_text)

print("Sample text:\n", resume_text[:500])  # preview first 500 chars
```

```python
from sentence_transformers import SentenceTransformer, util

model = SentenceTransformer('all-MiniLM-L6-v2')

job_description = """We are hiring a Salesforce Engineer with experience in Apex, Lightning, and integration with REST APIs."""
resume_embedding = model.encode(resume_text, convert_to_tensor=True)
jd_embedding = model.encode(job_description, convert_to_tensor=True)

similarity = util.pytorch_cos_sim(resume_embedding, jd_embedding).item()
print("Resume vs Job Description Similarity Score:", round(similarity, 2))
```
